```python
#Class property

class Item:
    discount=0.2 # 20%
    
    def __init__(self, name,price,quantity):
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def total_price(self):
        return self.price * self.quantity
    
    def apply_discount(self):
        self.price=self.price-Item.discount*self.price # Here discount from Item class is used
    
Iphone=Item('Iphone 15', 70000,1)
Iphone.apply_discount()
print(Iphone.__dict__)

Charger=Item('Charger', 1000,1)
Charger.apply_discount()
print(Charger.__dict__)
```

```python
#Class property
#always use self instead of class name in discount

class Item:
    discount=0.2 # 20%
    
    def __init__(self, name,price,quantity):
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def total_price(self):
        return self.price * self.quantity
    
    def apply_discount(self):
        self.price=self.price-self.discount*self.price # Here discount from self object is used
    
Iphone=Item('Iphone 15', 70000,1)
Iphone.apply_discount()
print(Iphone.__dict__)

Charger=Item('Charger', 1000,1)
Charger.discount=0.5
Charger.apply_discount()
print(Charger.__dict__)

# if no specific discount is given class discount will be used wheras if given it will override the class property
```

```python
#Getters --> get info

class ATM:
    __original_username='Ron'
    __original_password=1234
    phone=8344
    
    def __init__(self,username,password):
        self.username=username
        self.password=password
    
    #Getters --> get info
    
    def get_info(self):
        if (self.validate() == True):   # this if condition should be added so other users cannot access the info
            return self.__original_username,self.__original_password
        else:
            print('unauthorised')
        
    def validate(self):
        if(self.username==ATM.__original_username and self.password==ATM.__original_password):
            print('Access allowed')
            return True
        else:
            print('Access denied')
            

Ron=ATM('Ron',1234)
Ron.validate()
Ron.get_info()


John=ATM('John',1234)
John.get_info()

```

```python
#Setters --> change password

class ATM:
    __original_username='Ron'
    __original_password=1234
    phone=8344
    
    def __init__(self,username,password):
        self.username=username
        self.password=password
    
    #Getters --> get info
    
    def get_info(self):
        if (self.validate() == True):   # this if condition should be added so other users cannot access the info
            return self.__original_username,self.__original_password
        else:
            print('unauthorised')
    
    #setters
    def change_password(self,new_password):
        if (self.validate()):   # True condition is not required as that particular fn is already returning True or false only
            self.__original_password=new_password  # ALWAYS USE self while changing as using ATM/class will affect all objects
    
    
    def validate(self):
        if(self.username==ATM.__original_username and self.password==ATM.__original_password):
            print('Access allowed')
            return True
        else:
            print('Access denied')
            

Ron=ATM('Ron',1234)
Ron.validate()
print(Ron.get_info())  # print is reruired as the next get_info will overwrite this one, so printing to see the change
Ron.change_password(5678)
Ron.get_info()
```

```python
#Inheritance --> Parent class's property will be passed to the child class

# without inheritance:

class Item:
    
    def __init__(self, name,price = 0,quantity = 1):
        if price< 0 or quantity < 1:
            print('Invalid data') 
            return
        
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def total_price(self):
        return self.price * self.quantity
    
#to get more number of properties we can create a new class
# here default values for ram and rom is given to avoid the syntax(non-default argument follows default argument) error
class Phone:
    def __init__(self, name,price = 0,quantity = 1,ram='4gb',rom='64gb'):  
        if price< 0 or quantity < 1:
            print('Invalid data') 
            return
        
        self.name=name
        self.price=price
        self.quantity=quantity
        self.ram=ram
        self.rom=rom
    
    def total_price(self):
        return self.price * self.quantity

Iphone=Phone('Iphone 16',70000,-11,'6gb','64gb')

print(Iphone.__dict__)


# this is violating DRY principle as we are repeating the codes
```

```python
#with Inheritance:

class Funny:
    def __init__(self, name, price=0,quantity=1):
        assert price >= 0, f'Price {price} should be greater than or equal to zero'
        assert quantity > 0, f'Quantity {quantity} should be greater than zero'
                
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def total_amount(self):
        return self.price*self.quantity
    

class Dummy(Funny):
    def _init__(self, name, price=0,quantity=1, ram='4 gb', rom='64 gb'):
        super().__init__(name, price,quantity)
        self.ram=ram
        self.rom=rom

Fphone=Dummy('Fphone 16',60000,1,'16gb')
print(Fphone.__dict__)
```

```python
#Single level inheritance

class Parent:
    def __init__(self):
        self.property='Own house'
        
        
class Child(Parent):
    def __init__(self):
        super().__init__()
        self.bike='Duke'

Jacob=Parent()
Jake=Child()
print(Jake.bike)
print(Jake.property)  #wont work until super().__init__() is used
print(Jacob.property) 
#print(Jacob.bike)  wont work as Parent cannot inherit from Child
```

```python
#Multiple inheritance

class Father:
    def __init__(self):
        self.name='Jacob'
        self.property='Own house'

class Mother:
    def __init__(self):
        self.name='Julia'
        self.car='Tesla'      
        
class Child(Father,Mother):
    def __init__(self):
        super().__init__()   #if super keyword is used only the 1st Parent property are inherited
        self.bike='Duke'
        
Jacob=Father()
Julia=Mother()
Jake=Child()

print(Jacob.property)
print(Jake.property)
#print(Julia.property) --> cannot inherit as no established relation
#print(Jacob.car)  --> cannot inherit as no established relation
#print(Jake.car)  --> cannot inherit as parameter is from 2nd Parent
print(Julia.car)
```

```python
#Multiple inheritance using class name inplace of super keyword

class Father:
    def __init__(self):
        self.name='Jacob'
        self.property='Own house'

class Mother:
    def __init__(self):
        self.name='Julia'
        self.car='Tesla'      
        
class Child(Father,Mother):
    def __init__(self):
        Father.__init__(self) 
        Mother.__init__(self) 
        self.bike='Duke'
        
Jacob=Father()
Julia=Mother()
Jake=Child()

print(Jacob.property)
print(Jake.property)
#print(Julia.property)  # --> cannot inherit as no established relation
#print(Jacob.car)  --> cannot inherit as no established relation
print(Jake.car)  #can inherit even though parameter is from 2nd Parent as class/parent name is mentioned instead of super
print(Julia.car)
```

```python
#Heirarchial inheritance

class Parent:
    def __init__(self):
        self.name='Alec'
    
    def living(self):
        print(self.country)

        
class Child1(Parent):
    def __init__(self):
        
        self.name='Alexi'
        self.country='Russia'
        super().__init__()
        
class Child2(Parent):
    def __init__(self):
        
        self.name='Alex'
        self.country='USA'
        super().__init__()

Alec=Parent()
Alexi=Child1()
Alex=Child2()

Alexi.living()
Alex.living()


print(Alex.__dict__)
print(Alexi.__dict__)
```

```python
#Multi level inheritance using class name inplace of super keyword
class Grandfather:
    def __init__(self):
        self.name='John'
        self.property='Own house'

class Father(Grandfather):
    def __init__(self):
        super().__init__()
        self.name='Jacob'
        self.car='Tesla'      
        
class Child(Father):
    def __init__(self):
        super().__init__()
        self.name='Jake'
        self.bike='Duke'

John=Grandfather()        
Jacob=Father()
Jake=Child()

print(Jake.name)  # if print is not used only one value will be returned so use many codes print is used
print(Jake.bike)
print(Jake.car)
print(Jake.property)

print(Jacob.name)  
print(Jacob.car)
print(Jacob.property)

print(John.name)  
print(John.property)

# GF cant inherit car and bike, Father cant inherit bike
```

```python
class Parent:
    def play(self):
        print('Playing with parent')

class Child(Parent):
    name='Juju'
    def play(self):
        print('Playing with child')
    
Rand=Parent()
Rand.play()

Juju=Child()
Juju.play()

#Overridding
# IF both parent and child have same function with same name and parameters, child function will override parent's
```

```python
class Parent:
    def play(self,place):
        print('Playing with parent',place)

class Child(Parent):
    def play(self):
        print('Playing with child')
    
Rand=Parent()
Rand.play('Chennai')

Juju=Child()
Juju.play('Chennai') 

# Overloading
# IF both parent and child have same function with same name but different parameters, though Parent has a matching function
# child's command will priortize child's improper function and through error
#overloading is not supported by Python
```
