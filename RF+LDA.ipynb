```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('winequality-red.csv')
```

```python
#Splitting x-y

x=df.drop(columns=['quality'])
y=df['quality']
```

```python
y.value_counts()
```

```python
#Train-test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y)
```

```python
# Standardisation

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#Applying LDA

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda=LinearDiscriminantAnalysis()
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)
```

```python
#Model building
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params_grid={
'n_estimators': [10, 20, 30],# Number of trees in the forest
'max_features': ['sqrt', 'log2'],# Number of features to consider at each split
'max_depth': [None, 10, 20, 30],# Maximum depth of the tree
'min_samples_split': [2, 5, 10],# Minimum number of samples required to split an internal node
'min_samples_leaf': [1, 2, 4],# Minimum number of samples required to be at a leaf node
'bootstrap': [True, False]# Whether bootstrap samples are used when building trees
}
```

```python
grid_search=GridSearchCV(RandomForestClassifier(),param_grid=params_grid)
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_estimator_)
```

```python
new_RF=RandomForestClassifier(bootstrap=False, max_features='log2', min_samples_leaf=2,
                       min_samples_split=5, n_estimators=30,class_weight='balanced')
```

```python
new_RF.fit(x_train,y_train)
```

```python
y_pred=new_RF.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```
